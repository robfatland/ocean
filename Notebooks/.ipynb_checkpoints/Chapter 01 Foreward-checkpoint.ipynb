# Ocean Science Data Interpretation


This Jupyter Book is about *ocean science* by means of *data interpretation*.
There is a barrier between observatories that collect data and people 
constructively looking at and thinking about 
that data. This barrier consists
of a collection of technologies: Necessary to make progress but 
requiring time and effort to put into practice. 
The ambitious goal of this book is to 
streamline the data interpretation story and at the same time provide 
some of the keys to overcoming this technical barrier.


To begin let's start to build a model or framework for understanding
the ocean. Understand that the ocean is dark below a certain depth.
Imagine we are at the surface during the daytime, in bright sunlight; 
and we start to descend downwards.  With increasing depth the sunlight 
gradually grows fainter until it is nearly pitch black
down around a depth of perhaps 200 meters. This upper 200 meters 
of the ocean is the *photic zone*. This is where sunlight
is available as a direct energy source, particularly
to phytoplankton ('plant-like plankton') which form the base level 
of the ocean food web. Phytoplankton use available chemicals
and sunlight to store energy as carbohydrates, a first step 
in establishing an active ecosystem in the ocean.


From the photic zone as a starting point let's now jump ahead to
human activity, particularly our attempts to observe and understand
the ocean and more generally the earth system. In the US
the National Science Foundation has supported a program 
called OOI, which stands for Ocean Observatories Initiative. 
OOI is a collection of **arrays**,
that are in turn collections of sensors and support hardware. 
A single one of these arrays has an area of several 
thousand square kilometers.
There are seven arrays in the OOI program: Five in the 
northern hemisphere and two in the southern ocean. 
The southern ocean arrays operated for 
a few years starting in 2014 but have since been discontinued. 


The five northern hemisphere arrays are the Regional Cabled Array,
Global Station Papa Array, 
the Coastal Endurance Array, 
the Coastal Pioneer Array and the Global Irminger Sea Array.
Each has its own unique location, history and relevance to ocean science.



The Regional Cabled Array (RCA) is located off the Oregon coast and extends across the continental
shelf and several hundred kilometers further out into the deep ocean. 
This array, the RCA, is the starting
point for the data interpretation stories unfolded here. 
Among many other instruments, the RCA features three shallow profilers
that are designed to observe the photic zone in great detail.


- 500m depth: Outer edge of the continental shelf off of central Oregon
- 3100m depth: Further out at the base of the continental shelf off central Oregon
- 2100m depth: At the Juan de Fuca plate boundary, at the base of Axial Seamount


The shallow profiler **platform** is tethered to
the sea floor by means of two long cables. It is positively buoyant and rests 200 meters below 
the surface. The platform has both power and a data connection to shore. 


The **profiler** rests in a cradle on the **platform**. 
It is also positively buoyant. Platform and Profiler together look like this:

Under normal circumstances
this profiler is allowed to rise to near the surface (depth of approximately 10 meters) 
nine times each day. 
This is accomplished by means of a single cable on a winch. As the profiler 
ascends its "upward facing"
sensors acquire data. Once the profiler reaches the top of the profile 
it is winched back down again. 


Mean time in minutes for...

    
```
Ascent:    67
Descent:   45      (exception: local noon and midnight descents are about an hour longer)
Rest:      45
```


Ascent data are
considered more pristine; 
although pH and pCO2 are unique in that they are recorded on *descent*.
